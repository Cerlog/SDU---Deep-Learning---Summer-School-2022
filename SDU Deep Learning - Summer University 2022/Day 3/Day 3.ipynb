```python
df[['displacement', 'horsepower', 'weight']] = 1/(df[['displacement', 'horsepower', 'weight']])
```

```python
names = list(df.columns)
for i in range(1,9):
    fig, ax = plt.subplots()
    ax.set_ylabel(names[0])
    ax.set_xlabel(names[i])
    ax.scatter(df[names[0]],df[names[i]])
```

```python
#x = df[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin']]
#x = df['horsepower']
x = df[['horsepower', 'weight', 'acceleration', 'model_year', 'origin']] #
y = df['mpg']
x = sm.add_constant(x)

model = sm.OLS(y, x).fit()
predictions = model.predict(x)

print_model = model.summary()
print(print_model)
```
